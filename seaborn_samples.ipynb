```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

tips = sns.load_dataset("tips")
sns.set_style("ticks") # eksenlerdeki kesişim çigilerini ifade eder.
sns.set_context("notebook",font_scale = 1.5) # Grafikteki font'u (yazı boyutu) ayarlar.
sns.set_palette("cubehelix",1) #Renk paleti içinden (örneğin; cubehelix) numarası verilen renk seçilir(örneğin; 1)
g = sns.lmplot(x="tip", y="total_bill", hue="day", data=tips, aspect=2) #hue: belirtilen kolona göre datayı renkli olarak ayrıştırır. Aspect: figure için çerçeve oranı belirtir.
g = sns.FacetGrid(tips, col="time", row="smoker")
g = g.map(plt.hist, "total_bill")
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.scatter, "total_bill", "tip", edgecolor="w")
g = (g.set_axis_labels("Tip","Total bill(USD)").set(xlim=(0,10),ylim=(0,100)))

plt.title("title")
plt.show(g)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
```

```python
sns.boxplot(x="alive", y="age", hue="adult_male", data=titanic, palette="Set3", order=["yes", "no"]) # order ifadesindeki sıraya göre grafikte plot konumları sıralandırılır.
```

```python
sns.boxplot(data=iris,orient="h")
```

```python
g = sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=tips, kind="point", height=4, aspect=.7) # Farklı çeşitteki plot yöntemlerini kind olarak verip kullanma imkanı sunar.
```

```python
g = sns.catplot("alive", col="deck", col_wrap=4, data=titanic[titanic.deck.notnull()], kind="count", height=2.5, aspect=.8)
```

```python
 sns.violinplot(x="sex", y="age", data=titanic) #x yada y ekesenine konumladırılan kolona göre violin eksesi h yada v olarak otomatik ayarlanır.
```

```python
sns.violinplot(x="sex", y="age", data=titanic, scale="count", inner="stick") #count : datayı ölçeklendirir. inner : violin ekseninin türnü belirtir.
```

```python
sns.regplot(x="sepal_width", y="sepal_length",  data=iris, ci=68)
```

```python
sns.regplot(x="sepal_width", y="sepal_length",  data=iris, ci=100, truncate=False)# ci : destek vektörlerinin genişliğini ayarlar. truncate : line'ın çerçeveye fit olmasını sağlar.
```

```python
sns.regplot(x="sepal_width", y="sepal_length",  data=iris, truncate=False, robust=True)# robust regression eğrisine kendi hesapladığı destek vektörlerini çizer.
```

```python
corr = np.corrcoef(np.random.randn(10, 200))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True)
```
